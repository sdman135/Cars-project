## Deal with missing data

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# Function to count missing values in each column (True values are missing values)
def find_missing_values(dataframe):
    missing_data = dataframe.isnull()
    missing_data.head(5)

    for column in missing_data.columns.values.tolist():
        print(column)
        print (missing_data[column].value_counts())
        print("\n")
        
find_missing_values(df)
```

### Replace NaN with mean value

```python
# getting average of a column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
# Replace "NaN" by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# above code made into fuction to apply to multiple columns
def replace_nan_with_mean(column):
    mean_column = (column).astype("float").mean(axis=0)
    print(f"Average of {column.name}:", mean_column)
    column.replace(np.nan, mean_column, inplace=True)
    print(f"Replaced all NaN values, in column {column.name}, with the average value of:", mean_column) 
    
```

```python
replace_nan_with_mean(df["normalized-losses"])
```

```python
# mass apply function to replace nan values with the average value of the column
column_list = df["normalized-losses"],df['bore'],df['stroke'],df['horsepower'],df['peak-rpm']
for x in column_list:
    replace_nan_with_mean(x)
    print('\n\n')
```

### Replace NaN with the mode value

```python
df['num-of-doors'].value_counts()
```

```python
# find the most frequent value (mode)
df['num-of-doors'].value_counts().idxmax()

#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df
```

### Drop NaN value in target column

```python
# drop all rows with NA values in price column (my target value)
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
#double checking for missing values
find_missing_values(df)
```

## Convert data types to proper format

```python
# Checking data types of the dataframe before change
data_types1 = df.dtypes
data_types1
```

```python
# Converting the incorrect data type to the correct types
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
# Checking data types of the dataframe after change
data_types2 = df.dtypes
data_types2
```

```python
# Comparing data types of the dataframe before and after change
Compare_dtypes = data_types1 != data_types2
Compare_dtypes
```

```python
compare_dtypes_df = pd.DataFrame(data = (data_types1,data_types2, Compare_dtypes))
columns=["Columns","Orignal dtype","Changed dtype", "Did dtype change?"]
compare_dtypes_table = compare_dtypes_df.T.reset_index()
```

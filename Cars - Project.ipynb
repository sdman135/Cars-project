## Deal with missing data

```python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# Function to count missing values in each column (True values are missing values)
def find_missing_values(dataframe):
    missing_data = dataframe.isnull()

    for column in missing_data.columns.values.tolist():
        true_count = missing_data[column][missing_data[column] ==True].count()
        if true_count != 0:
#         print(column)
#         print (missing_data[column].value_counts())
            print(f"In column {column} there are {true_count} missing data points")
            print("\n")
        
find_missing_values(df)
```

### Replace NaN values with mean value for column

```python
# # getting average of a column
# avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
# print("Average of normalized-losses:", avg_norm_loss)
# # Replace "NaN" by mean value in "normalized-losses" column
# df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# above code made into function to apply to multiple columns
def replace_nan_with_mean(column):
    mean_column = (column).astype("float").mean(axis=0)
    print(f"Average of {column.name}:", mean_column)
    column.replace(np.nan, mean_column, inplace=True)
    print(f"Replaced all NaN values, in column {column.name}, with the average value of:", mean_column) 
    
```

```python
replace_nan_with_mean(df["normalized-losses"])
```

```python
# mass apply function to replace nan values with the average value of the column
column_list = df["normalized-losses"],df['bore'],df['stroke'],df['horsepower'],df['peak-rpm']
for x in column_list:
    replace_nan_with_mean(x)
    print('\n\n')
```

## Data Normalization

```python
# # replace (original value) with (original value)/(maximum value)
# df['length'] = df['length']/df['length'].max()
# df['width'] = df['width']/df['width'].max()
# df['height'] = df['height']/df['height'].max()

### above code in function form  ###

# scaling variable so the variable values range from 0 to 1
def normalize_column(columns_to_transform):
    for x in columns_to_transform:
        df[x] = df[x]/df[x].max()
    
    return df[columns_to_transform].head()
    
normalize_column(["length","width","height"])
```

## Indicator variable (dummy variable)

transforming categorical values in a column to numerical variable

```python
df.columns
```

```python
# get indicator variables and assign it to dummy variable
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
# change column names for ease of readability 
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```

```python
# get indicator variables of aspiration and assign it to data frame "dummy_variable_2"
dummy_variable_2 = pd.get_dummies(df['aspiration'])

# change column names for clarity
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

#merge the new dataframe to the original datafram
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "aspiration" from "df"
df.drop('aspiration', axis = 1, inplace=True)
```
